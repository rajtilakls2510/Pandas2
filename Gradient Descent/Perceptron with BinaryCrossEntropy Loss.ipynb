### In this notebook, we test how our handwritten Perceptron model works with Binary CrossEntropy Loss

**Note**: Our handwritten **myPerceptron** class is stored in **my_perceptron.py** which is in the same module as this notebook.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
%matplotlib inline
```

### Generating a random dataset with 1000 rows, 2 features and 2 classes

```python
X,y=make_blobs(n_samples=1000,n_features=2,centers=2)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

### Scaling the features

```python
scaler=StandardScaler()
```

```python
X=scaler.fit_transform(X)
```

### Train Test Split

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

## Importing our Perceptron class

```python
from my_perceptron import myPerceptron
```

```python
per=myPerceptron()
```

```python
per.fit(X_train,y_train,lr=0.01,epochs=100, show_epochs=10)
```

### Let's see how our cost reduced

```python
plt.plot(range(len(per.cost_history)),per.cost_history)
plt.xlabel('Epochs')
plt.ylabel('Cost')
plt.title("Cost at Epochs for Perceptron")
plt.show()
```

### The Accuracy Scores

```python
y_pred_train=per.predict(X_train)
y_pred_test=per.predict(X_test)
print("Training Accuracy: ",accuracy_score(y_pred_train,y_train))
print("Testing Accuracy: ",accuracy_score(y_pred_test,y_test))
```

### Let's Plot the Decision Boundary of our Perceptron model

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X,y,clf=per, legend=2)
```
