```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.tree import DecisionTreeRegressor
```

```python
data.head()
```

```python
X= data.drop(columns=['laptop_ID','Price_euros'])
y= data['Price_euros']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=1)
```

```python
ohe =OneHotEncoder()
ohe.fit(X[['Company','Product','TypeName','Inches','ScreenResolution','Cpu','Ram','Memory','Gpu','OpSys']])
```

```python
ohe.transform(X[['Company','Product','TypeName','Inches','ScreenResolution','Cpu','Ram','Memory','Gpu','OpSys']])
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_, sparse=False),['Company','Product','TypeName','Inches','ScreenResolution','Cpu','Ram','Memory','Gpu','OpSys']),
                                      remainder='passthrough')
```

```python
dtr = DecisionTreeRegressor()
```

```python
pipe = make_pipeline(column_trans, StandardScaler(), dtr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
pipe.steps[-1][1].get_depth()
```

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=i)
    pipe = make_pipeline(column_trans, StandardScaler(), dtr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
    print(f"Finished State: {i}")
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```
